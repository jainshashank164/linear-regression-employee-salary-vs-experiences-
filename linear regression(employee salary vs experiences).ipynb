```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("C://Users//JAIN//Downloads//Youtube_Machine_Learing_regression_03-main//Youtube_Machine_Learing_regression_03-main//dataset.csv")
dataset.head()
```

```python
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,1].values
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection  import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_predict=reg.predict(x_test)
y_predict
```

```python
y_test
```

```python
plt.scatter(x_train,y_train, color='red')
plt.plot(x_train,reg.predict(x_train), color='blue')
plt.title("linear regression salary vs experiences")
plt.xlabel("year of employee")
plt.ylabel("saleries of employee")
```

```python
plt.scatter(x_test,y_test, color='red')
plt.plot(x_test,reg.predict(x_test), color='blue')
plt.title("linear regression salary vs experiences")
plt.xlabel("year of employee")
plt.ylabel("saleries of employee")
```
